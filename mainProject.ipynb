```python
# function to generate the main menu of the program
def MainMenu():
   print("\n\tMain Menu:\n\
    1- Create a file that has Name, ID, and the total mark.\n\
    2- Show students that meet a specific criteria.\n\
    3- Add new student to the class.\n\
    4- Remove a student from the class.\n\
    5- Modify a student information.\n\
    6- Show the data of a student.\n\
    7- Terminate the program.\n")

# function to generate a menu if the user chooses (2) from the main menu
def secondMenu():
   print("\nWhat are you looking for?\n\
    1- Absence.\n\
    2- Midterm.\n\
    3- Classwork.\n\
    4- Final.\n\
    5- Cancel.\n")

#return user's choice if it is between 1-5
   while True:
      choice = input("Enter your choice (1-5): ")
      if (choice.isdigit()):
         choice = int(choice)
         if (1 <= choice <= 5):
            return choice
         else:
            print("please enter a number between 1-5 only.")
      else:
         print("please enter a number between 1-5 only.")

# function to create a file that has Name, ID, and the total mark. (choice == 1)
def generalData():
   infile = open("input_info", "r")
   outfile = open("output_info", "w")
   line = infile.readline()

#read each line and write to the new file
   while line != "":
      line = line.rstrip()
      line = line.rsplit()

      totalMark = float(line[3]) + float(line[4]) + float(line[5])
      outfile.write("%s " % line[1])
      outfile.write("%s " % line[0])
      outfile.write("%s \n" % str(totalMark))
      line = infile.readline()

   outfile.close()
   print("\n A new file (output_info) was created successfully, and it contains the following:")

   Lines = open("output_info", "r").readlines()
   for line in Lines:
      print(line.strip())

# function to show students that meet a specific criteria. (choice == 2)
def customSearch():
   choice = secondMenu()
   # Absence
   if (choice == 1):
      Lines = open("input_info", "r").readlines()
      for line in Lines:
         line = line.strip()
         line = line.rsplit()
         print(line[0], " ", line[1], " ", line[2])
   # Midterm
   elif (choice == 2):
      Lines = open("input_info", "r").readlines()
      for line in Lines:
         line = line.strip()
         line = line.rsplit()
         print(line[0], " ", line[1], " ", line[3])
   # Classwork
   elif (choice == 3):
      Lines = open("input_info", "r").readlines()
      for line in Lines:
         line = line.strip()
         line = line.rsplit()
         print(line[0], " ", line[1], " ", line[4])
   # Final
   elif (choice == 4):
      Lines = open("input_info", "r").readlines()
      for line in Lines:
         line = line.strip()
         line = line.rsplit()
         print(line[0], " ", line[1], " ", line[5])
   # Cancel
   elif (choice == 5):
      print("canceled by user.")
      return True

# function to add new student to the class. (choice == 3)
def addStudent():
   infile = open("input_info", "a")
   #write the new student ( new line ) to the file
   infile.write('\n')

   ID = input("Enter the student ID: ")
   infile.write("%s " % ID)

   Name = input("Enter student Name: ")
   infile.write("%s " % Name)

   Absences = input("Enter student Absences : ")
   infile.write("%s " % Absences)

   Midterm = input("Enter the student Midterm: ")
   infile.write("%s " % Midterm)

   Classwork = input("Enter the student Classwork: ")
   infile.write("%s " % Classwork)

   Final = input("Enter the student Final: ")
   infile.write("%s " % Final)

   infile.close()

   print("\nstudent added successfully.")

# function to remove student from the class. (choice == 4)
def removeStudent():
   student = int(input("Enter 1 to remove student by ID or 2 to remove by Name: "))

   #
   if student == 1:
      idNum = input("Enter the ID number: ")
      file = open("input_info", "r")
      lines = file.readlines()
      file.close()

      new_file = open("input_info", "w")
      for line in lines:
         if idNum in line:
            print(line)
            makeSure = input(" Press Enter to confirm removing: ")
            if makeSure == '':
               lines.remove(line)
               print("student removed successfully.")

      for i in lines:
         new_file.write(i)
         new_file.close


   elif student == 2:
      stuName = input("Enter the Student name: ")
      file = open("input_info", "r")
      lines = file.readlines()
      file.close()

      new_file = open("input_info", "w")
      nameList = []

      for line in lines:
         if stuName in line:
            nameList.append(line)

      if len(nameList) > 1:
         for i in range(len(nameList)):
            print(i + 1, nameList[i])

         studentsToDelete = int(input("which %s you want to delete? " % stuName))

         for i in range(len(nameList)):
            if studentsToDelete == i + 1:
               makeSure = input(" Press Enter to confirm removing: ")

               if makeSure == '':
                  lines.remove(nameList[i])
                  print("student removed successfully.")
      else:
         print(nameList[0])
         makeSure = input("Press Enter to confirm removing: ")
         if makeSure == '':
            lines.remove(nameList[0])
            print("student removed successfully.")

      for i in lines:
         new_file.write(i)
         new_file.close

   else:
      print("Sorry, invalid input. You must enter either 1 or 2 only!")

# function to modify student information. (choice == 5)
def modifyInfo():
   infile = open("input_info", 'r')
   lines = infile.readlines()
   infile.close()
   outfile = open("input_info", 'w')

   stuID = input("Enter student ID: ")
   n = 0
   for line in lines:
      line = line.split()
      if line[0] == stuID:
         print("%-10s : %-10s : %8s : %8s : %6s: %8s" % ("ID", "Name", "Absences", "Midterm", 'Classwork', "Final"))
         print("-" * 67)
         print("%-10s : %-10s : %8s : %8s : %8s : %8s" % (line[0], line[1], line[2], line[3], line[4], line[5]))
         print()

         print("Enter 1 to modify ID")
         print("Enter 2 to modify NAME")
         print("Enter 3 to modify absences")
         print("Enter 4 to modify Midterm")
         print("Enter 5 to modify Class work")
         print("Enter 6 to modify Final")

         dataToModify = input("Which information you want to modify? ")

         if dataToModify == "1":
            newdata = input("Enter the new ID: ")
            line[0] = newdata

            break
         elif dataToModify == "2":
            newdata = input("Enter the new NAME: ")
            line[1] = newdata

            break
         elif dataToModify == "3":
            newdata = input("Enter the new absences: ")
            line[2] = newdata

            break
         elif dataToModify == "4":
            newdata = input("Enter the new Midterm: ")
            line[3] = newdata

            break
         elif dataToModify == "5":
            newdata = input("Enter the new Class work: ")
            line[4] = newdata
            break
         elif dataToModify == "6":
            newdata = input("Enter the new Final: ")
            line[5] = newdata

            break


      else:
         print("Sorry, the ID is not found")

      n += 1

   newline = ''
   for ele in line:
      newline += ele + ' '
   newline += '\n'
   lines.insert(n, newline)
   lines.pop(n + 1)

   for line in lines:
      outfile.write(line)

   outfile.close()


# function to Show the data of a student. (choice == 6)
def ReadStudentsInfo():
   file = open("input_info", "r")
   print("%-10s  %-12s  %8s  %8s  %6s %8s" % ("ID", "Name", "Absences", "Midterm", 'Classwork', "Final"))
   print("-" * 67)
   Lines = file.readlines()
   for line in Lines:
      line = line.strip()
      line = line.split()
      print("%-10s  %-12s  %8s  %8s  %8s  %8s" % (line[0], line[1], line[2], line[3], line[4], line[5]))
   file.close()



#main of the program
print("\n--<<< Welcome to the Student Grade System >>>--")
while True:
   MainMenu()
   choice = input("Enter your choice (1-7): ")

# call specific function depends on the user's choice.
   if(choice.isdigit()):
      choice = int(choice)
      if (choice == 1):
         generalData()
      elif (choice == 2):
         customSearch()
      elif (choice == 3):
         addStudent()
      elif (choice == 4):
         removeStudent()
      elif (choice == 5):
         modifyInfo()
      elif (choice == 6):
         ReadStudentsInfo()
      elif (choice == 7):
         break
      else:
         print("please enter a number between 1-7 only.")
   else:
      print("please enter a number between 1-7 only.")
```
